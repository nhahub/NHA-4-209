### 3. Null Value Handling
Nulls were **not random** — they followed a logical pattern:

| Column Group | Null Count | Reason |
|---|---|---|
| DepTime, DepDelay... | ~120K | Flight was **Cancelled** — never departed |
| ArrTime, ArrDelay... | ~133K | Flight was **Cancelled or Diverted** — never arrived at destination |
| Tail_Number | ~26K | Data entry issue |

**Strategy — fill with `-1` (not drop):**
- `Cancelled` flights → all delay columns = `-1`
- `Diverted` flights → arrival columns only = `-1`
- `Tail_Number` → filled with `'UNKNOWN'`

> `-1` was chosen over `0` to distinguish *"no data"* from *"zero delay"*

```python
# Numeric col el nulls
delay_cols = ['DepTime', 'DepDelay', 'DepDelayMinutes', 'ArrTime', 'ArrDelay', 
              'ArrDelayMinutes', 'AirTime', 'ActualElapsedTime', 'ArrDel15',
              'DepDel15', 'TaxiOut', 'TaxiIn', 'WheelsOff', 'WheelsOn',
              'DepartureDelayGroups', 'ArrivalDelayGroups']

#  لو اتكنسلت هنعوض ب -1
cancelled_mask = df['Cancelled'] == True
df.loc[cancelled_mask, delay_cols] = df.loc[cancelled_mask, delay_cols].fillna(-1)


diverted_mask = df['Diverted'] == True
arr_cols = ['ArrTime', 'ArrDelay', 'ArrDelayMinutes', 'ArrDel15', 
            'ArrivalDelayGroups', 'TaxiIn', 'WheelsOn', 'ActualElapsedTime', 'AirTime']

df.loc[diverted_mask, arr_cols] = df.loc[diverted_mask, arr_cols].fillna(-1)

# 3. Tail_Number للكل
df['Tail_Number'] = df['Tail_Number'].fillna('UNKNOWN')


print(df.isnull().sum()[df.isnull().sum() > 0])
```

### 4. Flight Status Column
Instead of splitting into multiple DataFrames, we added a single `FlightStatus` column:
- `Completed` — flight departed and arrived normally
- `Cancelled` — flight never departed
- `Diverted` — flight landed at a different airport

This keeps the dataset unified and makes filtering easy in both Python and Power BI.

```python
df['FlightStatus'] = 'Completed'
df.loc[df['Cancelled'] == True, 'FlightStatus'] = 'Cancelled'
df.loc[df['Diverted'] == True, 'FlightStatus'] = 'Diverted'

completed = df[df['FlightStatus'] == 'Completed']
```

### 5. Data Type Fixes

| Column | Before | After | Reason |
|---|---|---|---|
| `FlightDate` | object | datetime64 | Enable time-based analysis |
| `DepTime`, `ArrTime`... | HHMM float | minutes from midnight | `715` → `435 min` — enables math operations |
| `Airline`, `Origin`... | object | category | Reduces memory significantly |

> Time columns like `DepTime` were stored as HHMM integers (e.g. `715` = 7:15 AM).  
> We converted them to **minutes from midnight** for accurate calculations.

```python
df['DepTime'].unique()
```

```python
def hhmm_to_minutes(val):
    """715.0 → 435 دقيقة من منتصف الليل"""
    if val == -1 or pd.isna(val):
        return -1
    val = int(val)
    hours = val // 100
    minutes = val % 100
    if hours == 24:
        hours = 0
    return hours * 60 + minutes

hhmm_cols = ['CRSDepTime', 'DepTime', 'CRSArrTime', 'ArrTime', 'WheelsOff', 'WheelsOn']

for col in hhmm_cols:
    df[col] = df[col].apply(hhmm_to_minutes)

# تحقق
print(df[['CRSDepTime', 'DepTime', 'ArrTime']].head())
print(df[['CRSDepTime', 'DepTime', 'ArrTime']].describe())
```

```python
# Datetime
df['FlightDate'] = pd.to_datetime(df['FlightDate'])

# Categories (بتوفر memory كتير مع الداتا الكبيرة دي)
cat_cols = ['Airline', 'Origin', 'Dest', 'FlightStatus',
            'OriginState', 'DestState', 'IATA_Code_Marketing_Airline',
            'IATA_Code_Operating_Airline', 'DepTimeBlk', 'ArrTimeBlk']

df[cat_cols] = df[cat_cols].astype('category')

# memory بعد التحويل
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1e9:.2f} GB")
```

## 6. Feature Engineering
New columns created to enrich analysis:

| Feature | Description |
|---|---|
| `Flight_Month`, `Flight_Day`, `Flight_DayOfWeek` | Extracted from `FlightDate` |
| `Is_Weekend` | 1 if Saturday or Sunday |
| `MonthName`, `DayName`, `Season` | Human-readable time labels |
| `DepTimeOfDay` | Red-Eye / Morning / Afternoon / Evening |
| `IsDelayed` | 1 if arrival delay > 15 min, -1 if not completed |
| `DelayCategory` | On Time / Minor / Moderate / Severe / Cancelled / Diverted |
| `RouteKey` | `"JFK→LAX"` format for route analysis |
| `DistanceCategory` | Short / Medium / Long / Ultra |
| `ScheduleAdherence` | Difference between scheduled and actual elapsed time |

```python
df['Flight_Month']     = df['FlightDate'].dt.month
df['Flight_Day']       = df['FlightDate'].dt.day
df['Flight_DayOfWeek'] = df['FlightDate'].dt.dayofweek
df['Is_Weekend']       = df['Flight_DayOfWeek'].isin([5, 6]).astype(int)

month_map = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
             7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
df['MonthName'] = df['Flight_Month'].map(month_map).astype('category')

day_map = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['DayName'] = df['Flight_DayOfWeek'].map(day_map).astype('category')

season_map = {12:'Winter',1:'Winter',2:'Winter',
              3:'Spring',4:'Spring',5:'Spring',
              6:'Summer',7:'Summer',8:'Summer',
              9:'Fall',10:'Fall',11:'Fall'}
df['Season'] = df['Flight_Month'].map(season_map).astype('category')

# ── من وقت الإقلاع — vectorized ──────────
conditions = [
    df['CRSDepTime'] == -1,
    df['CRSDepTime'] < 360,
    df['CRSDepTime'] < 720,
    df['CRSDepTime'] < 1020,
]
choices = ['Unknown', 'Red-Eye', 'Morning', 'Afternoon']
df['DepTimeOfDay'] = np.select(conditions, choices, default='Evening')
df['DepTimeOfDay'] = df['DepTimeOfDay'].astype('category')

# ── من الـ Delay — vectorized ─────────────
df['IsDelayed'] = (df['ArrDelay'] > 15).astype(int)
df.loc[df['FlightStatus'] != 'Completed', 'IsDelayed'] = -1

conditions = [
    df['FlightStatus'] == 'Cancelled',
    df['FlightStatus'] == 'Diverted',
    df['ArrDelay'] <= 0,
    df['ArrDelay'] <= 15,
    df['ArrDelay'] <= 60,
]
choices = ['Cancelled', 'Diverted', 'On Time', 'Minor', 'Moderate']
df['DelayCategory'] = np.select(conditions, choices, default='Severe')
df['DelayCategory'] = df['DelayCategory'].astype('category')

# ── من المطارات ──────────────────────────
df['RouteKey'] = df['Origin'].astype(str) + '→' + df['Dest'].astype(str)
df['RouteKey'] = df['RouteKey'].astype('category')

# ── من المسافة ───────────────────────────
df['DistanceCategory'] = pd.cut(
    df['Distance'],
    bins=[0, 500, 1000, 2000, 6000],
    labels=['Short', 'Medium', 'Long', 'Ultra']
).astype('category')

# ── Efficiency ───────────────────────────
mask = df['FlightStatus'] == 'Completed'
df.loc[mask, 'ScheduleAdherence'] = (
    df.loc[mask, 'CRSElapsedTime'] - df.loc[mask, 'ActualElapsedTime']
)
df['ScheduleAdherence'] = df['ScheduleAdherence'].fillna(-1)

print("Features added ")
print(df[['DelayCategory','DepTimeOfDay','Season','RouteKey','DistanceCategory']].head())
print(f"\nMemory: {df.memory_usage(deep=True).sum() / 1e9:.2f} GB")
```

## A report by excell

```python
big_airlines = df.groupby('Airline', observed=True).size()
big_airlines = big_airlines[big_airlines > 50000].index

# 2. report
airline_report = (df[df['Airline'].isin(big_airlines)]
                  .groupby('Airline', observed=True)
                  .agg(
                      Total_Flights=('FlightStatus', 'count'),
                      Completed=('FlightStatus', lambda x: (x == 'Completed').sum()),
                      Cancelled=('FlightStatus', lambda x: (x == 'Cancelled').sum()),
                      Diverted=('FlightStatus', lambda x: (x == 'Diverted').sum()),
                      Delayed=('IsDelayed', lambda x: (x == 1).sum()),
                      Avg_Arr_Delay=('ArrDelay', lambda x: x[x > 0].mean()),
                      Avg_Dep_Delay=('DepDelay', lambda x: x[x > 0].mean()),
                  )
                  .reset_index())

airline_report['OnTime_Rate'] = (airline_report['Completed'] - airline_report['Delayed']) / airline_report['Total_Flights'] * 100
airline_report['Cancel_Rate'] = airline_report['Cancelled'] / airline_report['Total_Flights'] * 100
airline_report['Delay_Rate']  = airline_report['Delayed']   / airline_report['Total_Flights'] * 100
airline_report['Score']       = (
    airline_report['OnTime_Rate'] * 0.5 +
    (100 - airline_report['Cancel_Rate']) * 0.3 +
    (100 - airline_report['Delay_Rate']) * 0.2
).round(1)

airline_report = airline_report.sort_values('Score', ascending=False)
print(airline_report[['Airline', 'Total_Flights', 'OnTime_Rate', 'Cancel_Rate', 'Score']].to_string(index=False))
```
